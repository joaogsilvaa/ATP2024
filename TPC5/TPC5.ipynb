### TPC5: Aplicação para Gerir um Cinema

Suponha que está a desenvolver uma aplicacão para gestão de um conjunto de salas de cinema de um centro comercial. 
Nesse centro comercial existem algumas salas de cinema (que poderão estar a exibir filmes ou não), cada sala tem uma determinada 
lotação, uma lista com a referência dos bilhetes vendidos (lugares ocupados; cada lugar é identificado por um número inteiro), e cada sala tem um filme associado.

Considera a seguinte sugestão para o modelo dos cinemas:
```
Cinema = [Sala]
Sala = [nlugares, Vendidos, filme]
nlugares = Int
filme = String 
Vendidos = [Int]
```
  
Que poderá ser usado num programa da seguinte forma:
```
sala1 = (150, [], "Twilight")
sala2 = (200, [], "Hannibal")
cinema1 = []
...
cinema1 = inserirSala(cinema1,sala1)
cinema1 = inserirSala(cinema1,sala2)
...
listar(cinema1)
...

if(disponivel(cinema1, "Twilight", 17 )):
  cinema1 = vendebilhete(cinema1, "Twilight", 17 )
...
listardisponibilidades(cinema1)
...
```

Especifique as funções utilizadas no exemplo:

1. `listar( cinema )` - que lista no monitor todos os filmes que estão em exibição nas salas do cinema passado como argumento;
2. `disponivel( cinema, filme, lugar )` - que dá como resultado **False** se o lugar lugar já estiver ocupado na sala onde o filme está a ser exibido e dará como resultado **True** se o inverso acontecer;
3. `vendebilhete( cinema, filme, lugar )` - que dá como resultado um novo cinema resultante de acrescentar o lugar à lista dos lugares ocupados, na sala onde está a ser exibido o filme;
4. `listardisponibilidades( cinema )` - que, para um dado cinema, lista no monitor para cada sala, o filme que está a ser exibido e o total de lugares disponíveis nessa sala (número de lugares na sala menos o número de lugares ocupados);
5. `inserirSala( cinema, sala )` - que acrescenta uma sala nova a um cinema (devendo verificar se a sala já existe);
6. Acrescente todas as outras funcionalidades que achar necessárias;
7. À semelhança do exercício 3, construa uma aplicação com um menu de interface para as operações.

````python
# Função para listar todos os filmes em exibição nas salas
def listar(cinema):
    for sala in cinema:
        print(f"Filme em exibição: {sala[2]}")

# Função para verificar se o lugar está disponível
def disponivel(cinema, filme, lugar):
    for sala in cinema:
        if sala[2] == filme:
            if lugar not in sala[1]:  # Verifica se o lugar está na lista de vendidos
                return True
            else:
                return False
    return False  # Caso o filme não exista

# Função para vender bilhete, adicionando o lugar à lista de vendidos
def vendebilhete(cinema, filme, lugar):
    for i, sala in enumerate(cinema):
        if sala[2] == filme:
            if disponivel(cinema, filme, lugar):  # Só vende se o lugar estiver disponível
                cinema[i] = (sala[0], sala[1] + [lugar], sala[2])  # Atualiza a lista de vendidos
                return cinema
            else:
                print(f"Lugar {lugar} já se encontra ocupado na sala {filme}")
                return cinema
    print(f"Filme {filme} não encontrado.")
    return cinema

# Função para listar as disponibilidades de lugares em cada sala
def listardisponibilidades(cinema):
    for sala in cinema:
        lugares_disponiveis = sala[0] - len(sala[1])  # Número de lugares disponíveis
        print(f"Filme: {sala[2]} - Lugares disponíveis: {lugares_disponiveis}")

# Função para inserir uma nova sala no cinema
def inserirSala(cinema, sala):
    for sala_existente in cinema:
        if sala_existente[2] == sala[2]:  # Verifica se o filme já está em exibição
            print(f"A sala com o filme {sala[2]} já existe!")
            return cinema
    cinema.append(sala)  # Caso contrário, adiciona a sala ao cinema
    return cinema

# Função para listar todas as salas com os detalhes
def listar_salas(cinema):
    for sala in cinema:
        print(f"Sala com {sala[0]} lugares, filme: {sala[2]}, vendidos: {sala[1]}")


def menu():
    cinema1 = []
    
    while True:
        print("\nMenu:")
        print("(1) Inserir Sala")
        print("(2) Listar Filmes em Exibição")
        print("(3) Verificar Disponibilidade")
        print("(4) Vender Bilhete")
        print("(5) Listar Disponibilidades")
        print("(0) Sair")

        opcao = input("Escolha uma opção: ")

        if opcao == '1':
            filme = input("Nome do filme: ")
            nlugares = int(input("Número de lugares na sala: "))
            sala = (nlugares, [], filme)
            cinema1 = inserirSala(cinema1, sala)
        
        elif opcao == '2':
            listar(cinema1)
        
        elif opcao == '3':
            filme = input("Nome do filme: ")
            lugar = int(input("Número do lugar: "))
            if disponivel(cinema1, filme, lugar):
                print(f"Lugar {lugar} disponível para o filme {filme}.")
            else:
                print(f"Lugar {lugar} já ocupado para o filme {filme}.")
        
        elif opcao == '4':
            filme = input("Nome do filme: ")
            lugar = int(input("Número do lugar: "))
            cinema1 = vendebilhete(cinema1, filme, lugar)
            print(f"Lugar {lugar} comprado! ")
        
        elif opcao == '5':
            listardisponibilidades(cinema1)
        
        elif opcao == '0':
            print("Saindo do sistema...")
            break

        else:
            print("Opção inválida! Tente novamente.")

# Chamar a função do menu
menu()
````
