### TPC6: Aplicação para gestão de alunos

Considere que o modelo do aluno e da turma têm a seguinte estrutura:

`aluno = (nome, id, [notaTPC, notaProj, notaTeste])`

`turma = [aluno]`

* Cria uma aplicação que coloca no monitor o seguinte menu de operações:
    - 1: Criar uma turma;
    - 2: Inserir um aluno na turma;
    - 3: Listar a turma;
    - 4: Consultar um aluno por id;
    - 8: Guardar a turma em ficheiro;
    - 9: Carregar uma turma dum ficheiro;
    - 0: Sair da aplicação
* No fim de executar a operação selecionada, a aplicação deverá colocar novamente o menu e pedir ao utilizador a opção para continuar;
* Utiliza a tua aplicação para criar uma turma com 5 alunos.

```python
import json

# Função para criar uma turma vazia
def criar_turma():
    return []

# Função para inserir um aluno na turma
def inserir_aluno(turma, nome, id, notas):
    aluno = (nome, id, notas)
    turma.append(aluno)
    return turma

# Função para listar todos os alunos na turma
def listar_turma(turma):
    if not turma:
        print("A turma está vazia.")
    for aluno in turma:
        print(f"Nome: {aluno[0]}, ID: {aluno[1]}, Notas: {aluno[2]}")

# Função para consultar um aluno pelo ID
def consultar_aluno(turma, id):
    for aluno in turma:
        if aluno[1] == id:
            print(f"Nome: {aluno[0]}, ID: {aluno[1]}, Notas: {aluno[2]}")
            return
    print("Aluno não encontrado.")

# Função para guardar a turma em um ficheiro JSON
def guardar_turma(turma, nome_ficheiro):
    with open(nome_ficheiro, 'w') as f:
        json.dump(turma, f, indent=4)
    print(f"Turma guardada em {nome_ficheiro}.")

# Função para carregar uma turma de um ficheiro JSON
def carregar_turma(nome_ficheiro):
    try:
        with open(nome_ficheiro, 'r') as f:
            turma = json.load(f)
        print(f"Turma carregada de {nome_ficheiro}.")
        return turma
    except FileNotFoundError:
        print("Ficheiro não encontrado.")
        return []

# Função para exibir o menu e interagir com o utilizador
def menu():
    turma = []
    while True:
        print("\nMenu:")
        print("1: Criar uma turma")
        print("2: Inserir um aluno na turma")
        print("3: Listar a turma")
        print("4: Consultar um aluno por ID")
        print("8: Guardar a turma em ficheiro")
        print("9: Carregar uma turma de um ficheiro")
        print("0: Sair da aplicação")
        
        opcao = input("Escolha uma opção: ")
        
        if opcao == '1':
            turma = criar_turma()
            print("Turma criada.")
        
        elif opcao == '2':
            nome = input("Nome do aluno: ")
            id = int(input("ID do aluno: "))
            notas = [float(input(f"Nota {i+1}: ")) for i in range(3)]
            turma = inserir_aluno(turma, nome, id, notas)
        
        elif opcao == '3':
            listar_turma(turma)
        
        elif opcao == '4':
            id = int(input("ID do aluno a consultar: "))
            consultar_aluno(turma, id)
        
        elif opcao == '8':
            nome_ficheiro = input("Nome do ficheiro para guardar a turma: ")
            guardar_turma(turma, nome_ficheiro)
        
        elif opcao == '9':
            nome_ficheiro = input("Nome do ficheiro para carregar a turma: ")
            turma = carregar_turma(nome_ficheiro)
        
        elif opcao == '0':
            print("Saindo da aplicação...")
            break
        
        else:
            print("Opção inválida! Tente novamente.")

# Chama a função menu para começar a aplicação

menu()
```
