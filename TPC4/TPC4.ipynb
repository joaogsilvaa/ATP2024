### TPC4: Aplicação para manipulação de listas de inteiros
- Crie uma aplicação em Python que coloca no monitor o seguinte menu:
    * (1) Criar Lista 
    * (2) Ler Lista
    * (3) Soma
    * (4) Média
    * (5) Maior
    * (6) Menor
    * (7) estaOrdenada por ordem crescente
    * (8) estaOrdenada por ordem decrescente
    * (9) Procura um elemento
    * (0) Sair
- O utilizador irá escolher uma das opções introduzindo o número correspondente;
- Se a opção não for sair, a aplicação executa a operação pretendida, apresenta o resultado e a seguir apresenta de novo o menu;
- Se a opção for sair, a aplicação termina colocando uma mensagem no monitor.

* No desenvolvimento da aplicação deverá ter em atenção o seguinte:
    - A aplicação terá uma variável interna para guardar uma lista de números;
    - Na opção 1, deverá ser criada uma lista de números aleatórios entre 1 e 100 que será guardada na variável interna;
    - Na opção 2, deverá ser criada uma lista com números introduzidos pelo utilizador, que será guardada na variável interna;
    - Nestas primeiras opções, se a variável interna já tiver uma lista, esta será sobreposta/apagada pela nova lista;
    - Na opção 3, será calculada a soma dos elementos na lista no momento;
    - Na opção 4, será calculada a média dos elementos na lista no momento;
    - Na opção 5, será calculado o maior elemento da lista no momento;
    - Na opção 6, será calculado o menor elemento da lista no momento;
    - Na opção 7, a aplicação deverá indicar (Sim/Não) se a lista está ordenada por ordem crescente;
    - Na opção 8, a aplicação deverá indicar (Sim/Não) se a lista está ordenada por ordem decrescente;
    - Na opção 9, a aplicação irá procurar um elemento na lista, se o encontrar deverá devolver a sua posição, devolverá -1 se o elemento não estiver na lista;
    - Se o utilizador selecionar a opção 0, a aplicação deverá terminar mostrando a lista que está nesse momento guardada.

```python
import random

# Função para verificar se a lista está ordenada de forma crescente
def esta_ordenada_crescente(lista):
    res = True
    for i in range(len(lista) - 1):
        if lista[i] > lista[i + 1]:
            res = False
    return res

def esta_ordenada_decrescente(lista):
    res = True
    for i in range(len(lista) - 1):
        if lista[i] < lista[i + 1]:
            res = False
    return res

# Função para procurar um elemento na lista e devolver sua posição ou -1
def procura_elemento(lista, elemento):
    if elemento in lista:
        return lista.index(elemento) + 1
    else:
        return -1

# Função principal
def menu():
    lista = []  # Variável para guardar a lista interna
    while True:
        # Exibir o menu
        print("\nMenu:")
        print("(1) Criar Lista")
        print("(2) Ler Lista")
        print("(3) Soma")
        print("(4) Média")
        print("(5) Maior")
        print("(6) Menor")
        print("(7) Está ordenada por ordem crescente?")
        print("(8) Está ordenada por ordem decrescente?")
        print("(9) Procurar um elemento")
        print("(0) Sair")

        opcao = input("Escolha uma opção: ")

        # Opção 1 - Criar Lista com números aleatórios entre 1 e 100
        if opcao == '1':
            lista = [random.randint(1, 100) for _ in range(10)]
            print("Lista criada aleatoriamente:", lista)

        # Opção 2 - Ler lista de números introduzidos pelo utilizador
        elif opcao == '2':
            lista = []
            n = int(input("Quantos números deseja introduzir? "))
            for _ in range(n):
                numero = int(input("Introduza um número: "))
                lista.append(numero)
            print("Lista introduzida:", lista)

        # Opção 3 - Soma dos elementos da lista
        elif opcao == '3':
            soma = sum(lista)
            print(f"Soma dos elementos da lista: {soma}")

        # Opção 4 - Média dos elementos da lista
        elif opcao == '4':
            if lista:
                media = sum(lista) / len(lista)
                print(f"Média dos elementos da lista: {media}")
            else:
                print("A lista está vazia!")

        # Opção 5 - Maior elemento da lista
        elif opcao == '5':
            if lista:
                maior = lista.sort()[-1]
                print(f"Maior elemento da lista: {maior}")
            else:
                print("A lista está vazia!")

        # Opção 6 - Menor elemento da lista
        elif opcao == '6':
            if lista:
                menor = lista.sort()[0]
                print(f"Menor elemento da lista: {menor}")
            else:
                print("A lista está vazia!")

        # Opção 7 - Verificar se a lista está ordenada por ordem crescente
        elif opcao == '7':
            if esta_ordenada_crescente(lista):
                print("A lista está ordenada por ordem crescente: Sim")
            else:
                print("A lista não está ordenada por ordem crescente: Não")

        # Opção 8 - Verificar se a lista está ordenada por ordem decrescente
        elif opcao == '8':
            if esta_ordenada_decrescente(lista):
                print("A lista está ordenada por ordem decrescente: Sim")
            else:
                print("A lista não está ordenada por ordem decrescente: Não")

        # Opção 9 - Procurar um elemento na lista
        elif opcao == '9':
            elemento = int(input("Digite o elemento a procurar: "))
            posicao = procura_elemento(lista, elemento) 
            if posicao != -1:
                print(f"Elemento encontrado na posição {posicao}")
            else:
                print("Elemento não encontrado na lista")

        # Opção 0 - Sair
        elif opcao == '0':
            print("Saindo... Lista final:", lista)
            break

        # Opção inválida
        else:
            print("Opção inválida, tente novamente.")

# Chamada da função principal para executar a aplicação
menu()
```
